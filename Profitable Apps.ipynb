# Create a dictionary to hold unique names and max reviews, call it max_reviews.


```python
max_reviews = {}
for row in gplaydata:
    name = row[0] #Assign the app name to variable `name`
    n_reviews = float(row[3]) #Assign the numbers of reviews column to `n_reviews`
    if name in max_reviews and max_reviews[name] < n_reviews: #if the var name is present as key in dict max_reviews
        #and if value of the key name is less than the new number of reviews if and only then assign the new
        # more number of reviews to as the new value to the dict with key name.
        max_reviews[name] = n_reviews
    if name not in max_reviews: #if var name is not present as key in dict max_reviews, create a new key-value
        #pair with the number of reviews as the value(which may or may not be overwritten).
        max_reviews[name] = n_reviews
```

```python
android_clean = []
already_added = []
for row in gplaydata:
    #name = row[0]
    #n_reviews = float(row[3])
    if float(row[3]) == max_reviews[row[0]] and row[0] not in already_added:
        android_clean.append(row)
        already_added.append(row[0])
```

After creating a dict with the highest number of reviews for each app name (without any duplicates, as dicts cannot have duplicates). We can remove the duplicates entries from the header-less complete dataset. To do this, we first create two empty lists, namely, the android_clean and already_added. The former will hold the new cleaned dataset with no duplicate entry and the latter will be used to keep track of the process.

Using a for loop we run a row by row iteration through the gplaydata(without the header list). In the loop we add an condition that appends the android_clean list only when the number of reviews is same as that present inside the dict. for that app name and when the app name is not in the already_added list.
This way only the highest number of reviews and unique entries are added to the android_clean to make it a new list of lists. Post which, the already_added is also appended with the lastest app name for the iteration thus keeping track of duplicity.

# To check if any non-english character app names are present

```python
def is_english(str):
    count = 0
    for i in str:
        if ord(i) > 127: # We know that non-eng char have ascii value greater than 127.
            count += 1
            if count > 3: # if there are more than 3 non-english char in the app names, we can than consider
                # the app name to be non-english, for simplicity and convenience.
                return False               
    return True
```

```python
# The above can also be done with a help of a function.

def freq_table(dataset, index):
    frq_tab = {}
    for row in dataset:
        if row[index] not in frq_tab:
            frq_tab[row[index]] = 0
        else:
            frq_tab[row[index]] += 1
    for key in frq_tab:  
        frq_tab[key] = round(((frq_tab[key]/len(dataset))*100), 2)
        
        # above code was simplified from the below code.
        '''
        frq_tab[key] /= len(dataset)
        frq_tab[key] *= 100 
        frq_tab[key] = round(frq_tab[key], 2)
        '''
        
    return frq_tab

plycatdict = freq_table(android_free, 1)
print(plycatdict, '\n')

appgendict = freq_table(apl_free, -5)
print(appgendict)
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index) # table is a dict. that needs to be sorted/ordered.
    table_display = [] # This list will be used to sort the contents into either ascending or descending.
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

The above function will show us the app category in a descending or ascending order by sorting through a list of tuples, that were created with a for loop on a dict. The sorted function does work on dict. as well , but only sorts the keys. Dict. are also considered as iterable as lists, tuples as shown and utilised in the code above. 

```python
display_table(android_free, 1)
print('\n')
display_table(apl_free, -5)
```

# Important to note
 It is crucial to understand at this point here, that the number of free apps for a category, does not necessarily have to mean, there are many number of users for that category. The descending list above only shows the number of such apps available as free in the app markets. However, the number of installs of an said app would indicate that the number of users or audience of that app or app category.


```python
appgendict = freq_table(apl_free, -5)

for genre in appgendict:
    total = 0
    len_genre = 0
    for row in apl_free:
        genre_app = row[-5]
        if genre_app == genre:
            n_ratings = float(row[5])
            total += n_ratings        
            len_genre += 1
    avg_ratngs = total/len_genre
    print(genre,':',avg_ratngs)
```

```python
plycatdict = freq_table(android_free, 1)

for cat in plycatdict: # first loop will go through each cat. in the cat. dict. one by one to -                         
    total_per_genre = 0 # -sum up all the installs
    len_genre = 0
    for row in android_free: 
        if row[1] == cat:
           # n_installs = row[5] # saves the number of installs
           # n_installs = n_installs.replace('+','') 
           # n_installs = float(n_installs.replace(',',''))
            total_per_genre += float(row[5].replace('+','').replace(',','')) #The str is replaced of commas -
            len_genre += 1 # -and + signs and converted to a float value.
    avg_installs_per_genre = total_per_genre/len_genre
    print(cat,':',avg_installs_per_genre)
```
